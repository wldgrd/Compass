# **Definindo Duas Funções Auxiliares**
    - rename_colums(dataframe) para tratar os nomes das colunas e deixar no padrao 'nome_da_coluna'
    - remove_duplicatas(dataframe) para remover as linhas duplicadas do data set

```python
def rename_columns(dataframe):
    df = dataframe.copy()
    title = lambda x: inflection.titleize(x)
    snakecase = lambda x: inflection.underscore(x)
    spaces = lambda x: x.replace(" ", "")
    cols_old = list(df.columns)
    cols_old = list(map(title, cols_old))
    cols_old = list(map(spaces, cols_old))
    cols_new = list(map(snakecase, cols_old))
    df.columns = cols_new
    return df
```

```python
df_aux = df1.loc[:, ['category', 'price']].groupby(['category']).mean().reset_index()
df_aux.columns = ['categoria', 'preco_medio']
df_aux = df_aux.sort_values('preco_medio', ascending = False).head(10)


plt.plot(df_aux['categoria'], df_aux['preco_medio'], marker ='o')
plt.title('Preço Médio Por Categoria')
plt.xlabel('Categoria')
plt.ylabel('Preço Médio ($)')
plt.xticks(rotation=90, ha='right')
plt.grid(visible = True, which = 'both')


for i, valor in enumerate(df_aux['preco_medio']):
    plt.text(df_aux['categoria'].iloc[i], valor, f'{valor:.2f}', fontsize = 8, ha = 'center', va = 'bottom')

plt.show()
```
