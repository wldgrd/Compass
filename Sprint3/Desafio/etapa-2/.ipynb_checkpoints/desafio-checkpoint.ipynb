# Definindo Duas Funções Auxiliares
    - rename_colums(dataframe) para tratar os nomes das colunas e deixar no padrao 'nome_da_coluna'
    - remove_duplicatas(dataframe) para remover as linhas duplicadas do data set

```python
def rename_columns(dataframe):
    df = dataframe.copy()
    title = lambda x: inflection.titleize(x)
    snakecase = lambda x: inflection.underscore(x)
    spaces = lambda x: x.replace(" ", "")
    cols_old = list(df.columns)
    cols_old = list(map(title, cols_old))
    cols_old = list(map(spaces, cols_old))
    cols_new = list(map(snakecase, cols_old))
    df.columns = cols_new
    return df
```

```python
def remove_duplicatas(dataframe):
    duplicado = df1.duplicated() #coluna escolhida para analisar duplicidade foi o nome do app
    linhas_nao_duplicadas = duplicado == False
    dataframe = dataframe.loc [ linhas_nao_duplicadas, :]
    return dataframe.reset_index(drop = True) ## Drop = True serve para derrubar a coluna com o índice antigo
```

**3. Faça um gráfico de pizza (pie chart) mostrando as categorias de apps existentes
		no dataset de acordo com a frequência em que elas aparecem.**

**-Listando as categorias e suas ocorrências  
   -Criando o gráfico**

```python
contagem_categorias = df1['category'].value_counts() #conta o número de ocorrências de cada categoria
```

```python
contagem_categorias
```

**- O ajuste do tamanho da figura foi feito por tentativa e erro até que chegasse a um resultado satisfatório em (45,15), bem como o ângulo de início foi testado para alguns valores e o 120 foi o que mais me agradou visualmente.  
-O parâmetro pctdistance foi adicionado para movimentar o valores percentuais para fora do gráfico a fim de facilitar a leitura.
-O parâmetro rotatelabels foi utilizado para que os nomes das categorias ficassem de forma radial, facilitando a leitura
-o parâmetro labeldistance foi utilizado para colocar os nomes das categorias para dentro do gráfico, pois julguei que assim a visualização fica mais limpa.**

```python
plt.figure(figsize=(45,15))
plt.pie( contagem_categorias, labels=contagem_categorias.index, 
        autopct = '%1.1f%%', startangle = 120, colors = plt.cm.Paired.colors, 
        rotatelabels = True, textprops={'fontsize': 8}, pctdistance = 1.05, labeldistance = 0.75)
plt.title('Frequência de Ocorrência das Categorias de Apps')
plt.tight_layout()
plt.show()
```
